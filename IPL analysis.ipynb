```python
# Find number of unique batsmen

unique_batsmen = ipl['batsman'].nunique()

print("Number of unique batsmen:", unique_batsmen)
```

```python
# Calculate total runs
total_runs = ipl['total_runs'].sum()

print("Total runs scored in the entire dataset:", total_runs)
```

```python
# Find the batsman with the highest runs

batsman_runs = ipl.groupby('batsman')['batsman_runs'].sum()
top_batsman = batsman_runs.sort_values(ascending=False).head(1)

print("Batsman with the most total runs:")
print(top_batsman)
```

```python
# Find the bowler who gave the most runs

bowler_runs = ipl.groupby('bowler')['total_runs'].sum()


most_expensive_bowler = bowler_runs.sort_values(ascending=False).head(1)

print("Bowler who gave the most runs:")
print(most_expensive_bowler)
```

```python
# Get the bowler with the most deliveries

balls_bowled = ipl['bowler'].value_counts()


top_bowler = balls_bowled.head(1)

print("Bowler who bowled the most deliveries (balls):")
print(top_bowler)
```

```python
# How many sixes were hit in total


total_sixes = ipl[ipl['batsman_runs'] == 6].shape[0]

print("Total number of sixes hit:", total_sixes)
```

```python
# how many fours were hit in total

total_fours = ipl[ipl['batsman_runs'] == 4].shape[0]

print("Total number of fours hit:", total_fours)
```

```python
# Which batsman hit the most sixes

sixes = ipl[ipl['batsman_runs'] == 6]


sixes_by_batsman = sixes['batsman'].value_counts()


top_sixer = sixes_by_batsman.head(1)

print("Batsman who hit the most sixes:")
print(top_sixer)
```

```python
#Which batsman hit the most fours

fours = ipl[ipl['batsman_runs'] == 4]

fours_by_batsman = fours['batsman'].value_counts()

top_fours_batsman = fours_by_batsman.head(1)

print("Batsman who hit the most fours:")
print(top_fours_batsman)
```

```python
#How many players were dismissed (got out) in total?
total_dismissals = ipl['player_dismissed'].notnull().sum()

print("Total number of players dismissed:", total_dismissals)
```

```python
# What is the most common way a player gets out (dismissal type)

dismissal_counts = ipl['dismissal_kind'].value_counts()


most_common_dismissal = dismissal_counts.head(1)

print("Most common dismissal type:")
print(most_common_dismissal)
```

```python
# How many dot balls were bowled in total

dot_balls = ipl[ipl['total_runs'] == 0]

# Count the number of such deliveries
total_dot_balls = dot_balls.shape[0]

print("Total number of dot balls bowled:", total_dot_balls)
```

```python
# Which over has the highest number of runs scored on average 

average_runs_per_over = ipl.groupby('over')['total_runs'].mean()


max_avg_over = average_runs_per_over.sort_values(ascending=False).head(1)

print("Over with the highest average runs scored:")
print(max_avg_over)
```

```python
# average runs scord per  over 

average_runs_per_over = ipl.groupby('over')['total_runs'].mean()


plt.figure(figsize=(10, 6))
average_runs_per_over.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Runs Scored Per Over (1 to 20)', fontsize=14)
plt.xlabel('Over Number', fontsize=12)
plt.ylabel('Average Runs', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# average runs scored per over(pie chart)

average_runs_per_over = ipl.groupby('over')['total_runs'].mean()

# pie chart
plt.figure(figsize=(8, 8))
plt.pie(average_runs_per_over, labels=average_runs_per_over.index,
        autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

plt.title('Average Runs Scored Per Over (Pie Chart)', fontsize=14)
plt.axis('equal')  
plt.show()
```

```python
#Top 10 Batsmen by Total Runs (Bar Chart) 

top_batsmen = ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)

top_batsmen.plot(kind='bar', figsize=(10,6), color='green', edgecolor='black')
plt.title('Top 10 Batsmen by Total Runs')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
 # Runs Per Over (Line Chart)

average_runs_per_over = ipl.groupby('over')['total_runs'].mean()

average_runs_per_over.plot(kind='line', marker='o', figsize=(10,5), color='purple')
plt.title('Average Runs per Over')
plt.xlabel('Over Number')
plt.ylabel('Average Runs')
plt.xticks(range(1, 21))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Top 10 Bowlers by Wickets (Bar Chart)

wickets = ipl[ipl['player_dismissed'].notnull()]
top_bowlers = wickets['bowler'].value_counts().head(10)

top_bowlers.plot(kind='bar', figsize=(10,6), color='red', edgecolor='black')
plt.title('Top 10 Bowlers by Wickets Taken')
plt.xlabel('Bowler')
plt.ylabel('Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
# top 10 batsmen bt balls faced 

balls_faced = ipl['batsman'].value_counts().head(10)

balls_faced.plot(kind='bar', figsize=(10,6), color='teal', edgecolor='black')
plt.title('Top 10 Batsmen by Balls Faced')
plt.xlabel('Batsman')
plt.ylabel('Balls Faced')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
#distrubution of extra runs

extras = ipl[['wide_runs', 'bye_runs', 'legbye_runs', 'noball_runs', 'penalty_runs']].sum()
extras = extras[extras > 0]

plt.figure(figsize=(7, 7))
plt.pie(extras, labels=extras.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Extra Runs', pad=30)
plt.axis('equal')
plt.show()
```

```python
# types of runs scored by batsmens

run_types = ipl['batsman_runs'].value_counts().sort_index()

plt.figure(figsize=(7, 7))
plt.pie(run_types, labels=run_types.index, autopct='%1.1f%%', startangle=140)
plt.title('Types of Runs Scored by Batsmen')
plt.axis('equal')
plt.show()
```
